```python
# EXTRACT DATA
# setup feature extractor with pathes to all data
fe = setup_feature_extractor()

# load doc_ids with their classification
man_class_file = join(DATA_PATH,"new_clean_manual_class_with_train.csv")
man_class = pd.read_csv(man_class_file, delimiter=',', header=0, quoting=1)

ids = list(man_class["doc_id"])
labels = list(man_class["class"])
train_ids, test_ids, train_labels, test_labels = train_test_split(ids, labels, test_size=0.3, random_state=42)

# create vectorizer and classifier
bow_vectorizer = TfidfVectorizer(token_pattern=r'(?u)\b\w\w\w+\b|©', min_df=1)
kwargs = {"penalty":'l2', "C":100, "fit_intercept":True,
          "intercept_scaling":1, "class_weight":"balanced",
          "solver":"liblinear", "max_iter":10, "tol":0.01}
bow_classifier = LogisticRegression(**kwargs)

# let the feature extractor train the bow classifiers on the train data only
for bc in fe.bow_classifiers:
    clean_train_data = bc.load_clean_data(list(train_ids))
    bc.fit_vectorizer(bow_vectorizer, clean_train_data, save_path=None)
    clean_num_data = bc.vectorizer.transform(clean_train_data)
    bc.train_classifier(bow_classifier, clean_num_data, np.array(train_labels), save_path=None)

c_date = str(strftime("%Y_%m_%d"))
#c_date = ""
train_data_path = join(DATA_PATH,"feature_values","train_"+c_date+"_cleanup.csv")
test_data_path = join(DATA_PATH,"feature_values","test_"+c_date+"_cleanup.csv")

fe.extract_features(doc_input=list(zip(train_ids,train_labels)) ,feature_file=train_data_path)
fe.extract_features(doc_input=list(zip(test_ids,test_labels)) ,feature_file=test_data_path)
```

```python
# LOAD DATA
c_date = str(strftime("%Y_%m_%d"))
c_date = "2017_08_10"
train_file = "train_"+c_date+"_cleanup.csv"
test_file = "test_"+c_date+"_cleanup.csv"

train_file = "train_2017_08_10_cleanup.csv"
test_file = "test_2017_08_10_cleanup.csv"

res_folder = join(RESULT_PATH,train_file[6:-4])
if(not(isdir(res_folder))):
    os.makedirs(res_folder)

train_data_path = join(DATA_PATH,"feature_values",train_file)
test_data_path = join(DATA_PATH,"feature_values",test_file)

train_data, train_labels, train_docs, column_names = load_data(train_data_path, norm=False)
test_data, test_labels, test_docs, column_names = load_data(test_data_path, norm=False)
test_data, val_data, test_labels, val_labels, test_docs, val_docs = train_test_split(test_data, test_labels, test_docs, test_size=0.5, random_state=42)

man_class_file = join(DATA_PATH,"new_clean_manual_class_with_train.csv")
man_class = pd.read_csv(man_class_file, delimiter=',', header=0, quoting=1)
man_class = man_class.set_index("doc_id")

cat_types = [None,None,None]
for i,doc_type in enumerate([train_docs, test_docs, val_docs]):
    print(len(doc_type))
    cat_types[i]  = np.zeros(len(doc_type))
    for j,d_id in enumerate(doc_type):
        cat_types[i][j] = man_class.loc[d_id]["category"]
train_cat, test_cat, val_cat = cat_types

for i,doc_type in enumerate([train_labels, test_labels, val_labels]):
    print(np.sum(doc_type==1)/len(doc_type))

print(column_names)
```

```python
# SCALING LOG OR MINMAX
scaling = "log"

# prepare complete data
complete_data = np.concatenate([train_data,test_data,val_data],axis=0)
test_offset = len(train_data)
val_offset = len(train_data)+len(test_data)
# do scaling
if(scaling=="log"):
    complete_scaled_data = log_transform(complete_data)
elif(scaling=="minmax"):
    complete_scaled_data = minmax_scaling(complete_data)
else:
    print("Not a valid scaler! No scaling done!")
    complete_scaled_data = complete_data
scaled_train_data = complete_scaled_data[0:test_offset,:]
scaled_test_data = complete_scaled_data[test_offset:val_offset,:]
scaled_val_data = complete_scaled_data[val_offset:,:]
```

```python
# LOGISTIC REGRESSION
kwargs = {"penalty":'l2', "C":0.1, "fit_intercept":True,
          "intercept_scaling":1, "class_weight":None,
          "solver":"liblinear", "max_iter":1000, "tol":0.03}
lr = LogisticRegression(**kwargs)

lr_param_dist = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                 "solver": ["liblinear"],
                 "class_weight": [None, "balanced"]}

lr_p_keys = ["C", "solver", "class_weight"]
lr_p_keys_abrev = ["C", "solver", "c-weight"]
```

```python
# FOREST
kwargs = {"n_estimators":50, "criterion":'entropy', "max_depth":16,
          "min_samples_leaf":8, "max_features":8, "bootstrap":True,
          "n_jobs":1, "class_weight":"balanced"}
rf = RandomForestClassifier(**kwargs)

rf_param_dist = {"n_estimators":list(np.arange(50)+30),
                 "max_depth": list(np.arange(38)+2)+[None],
                 "max_features": [2,3,4,5,6,7,8,10,15,20,25],
                 "min_samples_leaf": [2,3,4,5,6,7,8,10,12,14,16]}

rf_p_keys = ["n_estimators", "max_depth", "max_features", "min_samples_leaf"]
rf_p_keys_abrev = ["trees", "depth", "feat", "msl"]
```

```python
# SVM
kwargs = {"C":1.0, "kernel":'rbf', "degree":3, "gamma":'auto',"coef0":0.0,
          "probability":True, "class_weight":"balanced"}
svm = SVC(kwargs)

svm_param_dist = {"C":[0.01,0.01,0.5,1,10,100,1000,10000],
                  "kernel": ["linear", "poly", "rbf", "sigmoid"],
                  "gamma": [0.00001,0.0001,0.001,0.01,0.1,1,2,5,10],
                  "probability": [True],
                  "class_weight":["balanced",None]}

svm_p_keys = ["C", "kernel", "gamma", "probability", "class_weight"]
svm_p_keys_abrev = ["C", "kernel", "gamma", "prob", "c-weight"]
```

```python
# DECISION TREE
kwargs = {"criterion":'gini', "splitter":'random', "max_depth":8,
          "min_samples_split":2, "min_samples_leaf":16, "max_features":16,
          "min_impurity_split":1e-07, "class_weight":None}
dt = DecisionTreeClassifier(kwargs)

dt_param_dist = {"max_depth": list(np.arange(38)+2)+[None],
                 "max_features": [2,4,8,16,32,64],
                 "min_samples_leaf": [2,3,4,5,6,7,8,10,12,14,16],
                 "criterion":['entropy','gini'],
                 "splitter":['best','random'],
                 "class_weight":["balanced",None]}

dt_p_keys = ["max_depth", "max_features", "min_samples_leaf", "criterion", "splitter", "class_weight"]
dt_p_keys_abrev = ["depth", "feat", "msl", "crit", "split", "c-weight"]
```

```python
# MULTI LAYER PERCEPTRON
layer_params = {"kernel_initializer":"glorot_uniform", "activation":"sigmoid"}
compile_params = {"loss":"mean_squared_error", "optimizer":"rmsprop", "metrics":["accuracy"]}
train_params = {"epochs":30, "batch_size":16, "verbose":0}

mlp = Keras_Dense_MLP(neuron_layers=[len(train_data[0]), 1069, 199, 1], layer_params=layer_params, compile_params=compile_params, **train_params)

def random_layer_gen():
    fc = len(scaled_train_data[0])
    layers = [fc]
    layers.append(np.random.randint(fc*2,fc*15))
    add_alyer = np.random.choice([True,False])
    if(add_alyer):
        layers.append(np.random.randint(fc*2,fc*10))
    layers.append(1)
    return layers

mlp_param_dist = {"epochs": [10*i for i in range(3,4)],
                  "batch_size": [np.power(2,i) for i in range(3,7)],
                  "neuron_layers": random_layer_gen,
                  "optimizer":["sgd","rmsprop"],
                  "loss":["mean_squared_error","mean_absolute_error"]}

mlp_p_keys = ["epochs", "batch_size", "neuron_layers", "optimizer","loss"]
mlp_p_keys_abrev = ["epochs", "b-size", "layers", "opt","loss"]
```

```python
# PARAMETER OPTIMIZATION
classifier = lr
classifier_name = "lr"
param_dist = lr_param_dist
p_keys = lr_p_keys
p_keys_abrev = lr_p_keys_abrev
search_option = "grid"
sort_by = "f_beta"
use_deepcopy = True
```

```python
if(search_option=="random"):
    res, header = random_search_bin(classifier=classifier,
        n_models=200,
        params=param_dist,
        train_data=scaled_train_data,
        train_labels=train_labels,
        test_data=scaled_test_data,
        test_labels=test_labels,
        sort_by=sort_by,
        use_deepcopy = use_deepcopy,
        results_file=join(res_folder,classifier_name+"_"+search_option + "_opt.txt"))
elif(search_option=="grid"):
    res, header = grid_search_bin(classifier=classifier,
        params=param_dist,
        train_data=scaled_train_data,
        train_labels=train_labels,
        test_data=scaled_test_data,
        test_labels=test_labels,
        sort_by="f_beta",
        use_deepcopy = use_deepcopy,
        results_file=join(res_folder,classifier_name+"_"+search_option + "_opt.txt"))
else:
    print("Wrong search option!")
```

```python
from prettytable import PrettyTable
res_tab = PrettyTable(header)
model_tab = PrettyTable(p_keys_abrev)

res_sort = sorted(res, key=lambda x: x[0], reverse=True)
for r in res_sort[0:20]:
    res_tab.add_row(list(r[2:]))
    model_tab.add_row([r[0][k] for k in p_keys])
print(res_tab.get_string())
print(model_tab.get_string())
```

```python
# CLASSIFY
classifier = mlp
#classifier.load_model(join(res_folder,"models","mlp_best_fbeta.model"))
#classifier = joblib.load(join(res_folder,"models","mlp_best_fbeta.model"))
classifier.fit(scaled_train_data, train_labels)
#classifier = res_sort[0][1]

n_docs = len(val_labels)
n_copy = np.sum(val_labels)
n_not = np.sum(val_labels==0)

print("total: %d\tcopyright: %d\tnot_protected: %d"%(n_docs,n_copy,n_not))

probs = classifier.predict_proba(scaled_train_data)[:,1]
preds = probs>0.5

print("train_acc: %.4f"%(accuracy_score(train_labels,preds),))

probs = classifier.predict_proba(scaled_val_data)[:,1]
preds = probs>0.5

print("acc: %.4f"%(accuracy_score(val_labels,preds),))
print("f1: %.4f"%(f1_score(val_labels,preds),))
print("fbeta: %.4f"%(fbeta_score(val_labels,preds,1.7),))
print("precision: %.4f"%(precision_score(val_labels,preds),))
print("recall: %.4f"%(recall_score(val_labels,preds),))
conf_mat = confusion_matrix(val_labels,preds)
plot_confusion_matrix(conf_mat,["not_protected","copyright"],True)

fpr, tpr, _ = roc_curve(val_labels, probs)
auc_roc = auc(fpr, tpr)
plot_curve(fpr,tpr,'False Positive Rate','True Positive Rate','area = %0.3f'%(auc_roc,),'ROC curve')

fnr, tnr, _ = roc_curve(~(val_labels.astype(bool)), 1-probs)
inv_auc_roc = auc(fnr, tnr)
plot_curve(fnr,tnr,'False Negative Rate','True Negative Rate','area = %0.3f'%(inv_auc_roc,),'Inverse ROC curve')

precision, recall, thresholds = precision_recall_curve(val_labels, probs)
avg_pr = average_precision_score(val_labels, probs)
plot_curve(recall,precision,'Recall','Precision','avg_pr = %0.3f'%(avg_pr,),'PR curve')
```

```python
# CATEGORY ANALYSIS
cat_header = ["category", "cnt", "acc", "tn", "fp", "fn", "tp"]
cat_tab = PrettyTable(cat_header)
for k,v in cat_id_to_des.items():
    cat_res = [v]
    cat_preds = preds[val_cat==k]
    cat_labels = val_labels[val_cat==k]
    cat_res.append("%d  %.3f"%(len(cat_preds),len(cat_preds)/len(preds)))
    cat_res.append("%.4f"%(accuracy_score(cat_labels, cat_preds),))
    c_mat = confusion_matrix(cat_labels,cat_preds).ravel()
    if(len(c_mat)==1):
        if(cat_labels[0]==0):
            c_mat = list(c_mat) + [0,0,0]
        else:
            c_mat = [0,0,0] + list(c_mat)
    cat_res.extend([str(val) for val in c_mat])
    cat_tab.add_row(cat_res)
print(cat_tab.get_string())
```
